# Sensitivity Analysis Demo

This draws heavily from the tutorial in the SALib package documentation:
https://salib.readthedocs.io/en/latest/user_guide/basics.html

#### Question: 
What's the purpose of a sensitivity analysis?

#### Answer: 
To determine how the uncertainty in a model's output can be attributed to the uncertainty in its inputs. This is important because it helps us hone our understanding of which parts of a model truly matter to the end result. We can't dedicate resources towards making every parameter perfect, so let's try to identify the most impactful ones and put our efforts towards improving those.

We'll accomplish this with the python package SALib, and the Sobol method. The Sobol method is a variance based sensitivity analysis technique, comparing the variance in results by differences in the parameter values.

## Set Up
If you haven't already installed SALib you can do so via conda in the terminal, or with pip. Remove the comment if you need to install it still.

```python
# pip install SALib
# import some commonly used packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# SALib functions
from SALib.sample import saltelli # this creates samples via the saltelli method
from SALib.analyze import sobol # this applies the sobol sensitivity analysis to our data
```

Let's make up a function to test the sobol method. We also need to define our model inputs, which we do so in the dictionary "problem".

$$y = x_1 + x_1 * x_2^2 + e^{x_3}$$

```python
def model(x1, x2, x3):
    function = x1 + x1 * x2**2 + np.exp(x3)
    return function

problem = {
    'num_vars': 3, # the number of variables in our function
    'names': ['x1','x2','x3'], # the names of the variables
    'bounds': [[-10, 10], # lower and upper bounds for x1
               [-3.14159265359, 3.14159265359], # lower and upper bounds for x2
               [-3.14159265359, 3.14159265359]] # lower and upper bounds for x3
}
```

Now we need to generate samples to run through our model. We need to use the saltelli sampler since we're running a Sobol analysis (there's some background math that I don't understand that requires this specific number), but the saltelli method also does a better job of covering the sample space than a random sample would. 

```python
param_values = saltelli.sample(problem, 2**10)
```

Quick example of sampling differences:

```python
# example dict to draw samples from
saltelli_samples = {
    'num_vars': 2,
    'names': ['x1','x2'],
    'bounds': [[-np.pi, np.pi],
               [-np.pi, np.pi]]
}

saltelli_samples = saltelli.sample(saltelli_samples, 2**3)
num_saltelli_samples = saltelli_samples.shape[0] # extract the total number of saltelli samples to use as the same number for the random samples

random_samples = np.random.uniform(low=-np.pi, high=np.pi, size=(num_saltelli_samples,2)) # create a uniform distribution to draw samples from

fig, axes = plt.subplots(1,2,figsize=(8,4))

axes[0].scatter(saltelli_samples[:,0], saltelli_samples[:,1], alpha=0.7)
axes[0].set_title('Saltelli Samples')
axes[0].set_xlabel('x1')
axes[0].set_ylabel('x2')

axes[1].scatter(random_samples[:,0], random_samples[:,1], alpha=0.7)
axes[1].set_title('Random Samples')
axes[1].set_xlabel('x1')
axes[1].set_ylabel('x2')
plt.tight_layout()
plt.show()
```

The saltelli sampler repeats some values but generally covers the sample space better.

## Evaluate Model
Run the model for all the generated samples. Normally we would save the results to an external file, but the number of samples is small enough here to just keep it in out environment.

```python
Y = np.zeros(param_values.shape[0]) # make an empty array to store the model results in

for i, X in enumerate(param_values): # enumerate is nice because the "i" corresponds to the index while the "X" corresponds to the actual value
    Y[i] = model(X[0], X[1], X[2]) # loop through each one of the sampled parameters and evaluate the model
```

## Sobol Analysis
Compute the sensitivity indices. These are measures of how influential a parameter is to the model.
- First order: the contribution to output variance by the parameter input
- Second order: the contribution to output variance caused by the interaction of two model inputs
- Total order: the contribution to output variance from an input's first order and all higher order interactions

```python
Si = sobol.analyze(problem, Y, print_to_console=True) # Si is a python dict
```

```python
Si['ST']
```

```python
plt.figure(figsize=(14,4))
Si.plot() # SALib includes basic plotting abilities that can be adjusted as a regular matplotlib axes
plt.tight_layout()
plt.show()
```

So these results tell us that $x_1$ is the most influential parameter, stemming from its own first order influence and the interactive effect it has with $x_2$. When the confidence intervals cross the zero line this means we're pretty sure the parameter has no influence for that specific indices. If the CIs are very wide we probably need to run the analysis with a larger sample size.

```python
total_Si, first_Si, second_Si = Si.to_df() # can convert Si to a pandas dataframe for easier analysis
# this makes more in-depth plotting easier, imo
```
